## 2. Variables & Basic Data Types
A variable is simply a name that stores a value. You create them by assigning a value using `=`.

**Common Types:**
* `str`: Text (e.g., "Hello")
* `int`: Whole numbers (e.g., 42)
* `float`: Decimal numbers (e.g., 3.142)
* `bool`: Logical values (True/False)

```python
name = "Jon"
age = 45
is_active = True
height = 1.82

print(f"Name: {name}, Type: {type(name)}")
```

## 3. Collections
These are the containers you'll use constantly in AI pipelines.

* **List:** Ordered, changeable `["a", "b"]`
* **Tuple:** Ordered, not changeable `(1.2, 3.4)`
* **Dictionary:** Key/value pairs `{"key": "value"}`
* **Set:** Unique values `{1, 2, 3}`

```python
languages = ["Python", "Java", "RPG"]
user = {
    "name": "Jon",
    "role": "Core Banking IT",
    "active": True
}

print(languages[0])
print(user["role"])
```

## 4. Conditional Logic
This lets your program make decisions. Indentation matters — four spaces is the standard.

```python
age = 18

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
```

## 5. Loops
Loops repeat actions — critical for processing files, data, or records.

```python
tools = ["FAISS", "Ollama", "Dash"]

for tool in tools:
    print("Using:", tool)

count = 3
while count > 0:
    print("Countdown:", count)
    count -= 1
```

## 6. Functions
A function is reusable code with a name. This is the backbone of structuring larger programs.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Jon"))
```
