## 3. Collections
[cite_start]These are the containers you'll use constantly in AI pipelines[cite: 27].

* **Tuple:** Ordered, not changeable `(1.2, 3.4)`

## 4. Conditional Logic
This lets your program make decisions. [cite_start]Indentation matters — four spaces is the standard[cite: 49].

```python
age = 18

if age >= 18:
    print("Adult")
elif age >= 13:
    print("Teenager")
else:
    print("Child")
```

## 5. Loops
[cite_start]Loops repeat actions — critical for processing files, data, or records[cite: 51].

```python
tools = ["FAISS", "Ollama", "Dash"]

for tool in tools:
    print("Using:", tool)

count = 3
while count > 0:
    print("Countdown:", count)
    count -= 1
```

## 6. Functions
A function is reusable code with a name. [cite_start]This is the backbone of structuring larger programs[cite: 62, 66].

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Jon"))
```
