````python
# Run this cell to display the Submit button
import requests
import ipywidgets as widgets
from IPython.display import display
import os

SUBMIT_URL = "https://script.google.com/macros/s/AKfycbyZD3jvyZeluWLDFc-NrmQyKsSPyL1_YmjrRezy2mfXmZVfwDXCQc70nOivp_YYARaQ/exec"

def get_user_email():
    """Get user's email from Colab authentication"""
    try:
        from google.colab import auth
        auth.authenticate_user()
        import google.auth
        creds, _ = google.auth.default()
        return creds.service_account_email if hasattr(creds, 'service_account_email') else 'colab_user'
    except:
        return 'unknown_user'

def submit_notebook(button):
    button.disabled = True
    button.description = "Submitting..."
    
    try:
        # Save notebook first
        try:
            from google.colab import _message
            _message.blocking_request('save')
        except:
            pass
        
        # Get user email
        user_email = get_user_email()
        
        # Find and read notebook
        notebook_path = None
        for f in os.listdir('.'):
            if f.endswith('.ipynb'):
                notebook_path = f
                break
        
        if not notebook_path:
            notebook_path = 'Module1_Assessment.ipynb'
        
        with open(notebook_path, 'r') as f:
            notebook_content = f.read()
        
        # Submit
        response = requests.post(
            SUBMIT_URL,
            json={
                'studentName': user_email,
                'studentEmail': user_email,
                'notebookContent': notebook_content
            }
        )
        
        result = response.json()
        
        if result.get('success'):
            print("=" * 50)
            print("SUBMISSION SUCCESSFUL!")
            print("=" * 50)
            print(f"\\nYour assessment has been submitted for grading.")
        else:
            print("SUBMISSION FAILED:", result.get('message'))
            
    except Exception as e:
        print(f"ERROR: {e}")
    
    button.disabled = False
    button.description = "Submit Assessment"

# Create and display button
submit_btn = widgets.Button(
    description='Submit Assessment',
    button_style='success',
    style={'button_color': '#28a745', 'font_weight': 'bold'}
)
submit_btn.on_click(submit_notebook)
display(submit_btn)
````
